```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import linregress

# Load data
forecasts_sun = np.load("../data/forecasts_sun.npy")
forecasts_rain = np.load("../data/forecasts_rain.npy")
measures_sun = np.load("../data/measures_sun.npy")
measures_rain = np.load("../data/measures_rain.npy")

# Function to plot and improve visualization
def plot_forecast_vs_measured(forecast, measured, title, xlabel, ylabel, filename):
    plt.figure(figsize=(10, 6))  # Bigger figure size
    plt.plot(forecast, c="b", linewidth=2, label="Forecast")  # Thicker lines
    plt.plot(measured, c="r", linewidth=2, label="Measured")  

    # Titles and labels
    plt.title(title, fontsize=18, fontweight="bold")
    plt.xlabel(xlabel, fontsize=16, fontweight="bold")
    plt.ylabel(ylabel, fontsize=16, fontweight="bold")

    # Grid and ticks
    plt.grid(True, linestyle="--", alpha=0.6)
    plt.xticks(fontsize=14)
    plt.yticks(fontsize=14)

    # Legend styling
    plt.legend(fontsize=14, loc="best", frameon=True, edgecolor="black")

    # Save and show
    plt.savefig(filename, dpi=300, bbox_inches="tight")
    plt.show()

# **Sun Data**
# Daily plot
plot_forecast_vs_measured(
    forecasts_sun[209:233], measures_sun[209:233], 
    "Daily Temperature Forecast vs Measured (Sun)", 
    "Hours since start of recording", "Temperature (K)", 
    "forecast_vs_real_sun_daily.png"
)
print(linregress(forecasts_sun[209:233], measures_sun[209:233]))

# Weekly plot
plot_forecast_vs_measured(
    forecasts_sun[209:377], measures_sun[209:377], 
    "Weekly Temperature Forecast vs Measured (Sun)", 
    "Hours since start of recording", "Temperature (K)", 
    "forecast_vs_real_sun_weekly.png"
)
print(linregress(forecasts_sun[209:377], measures_sun[209:377]))

# **Rain Data**
# Daily plot
plot_forecast_vs_measured(
    forecasts_rain[209:233], measures_rain[209:233], 
    "Daily Precipitation Forecast vs Measured", 
    "Hours since start of recording", "Precipitation (mm)", 
    "forecast_vs_real_rain_daily.png"
)
print(linregress(forecasts_rain[209:233], measures_rain[209:233]))

# Weekly plot
plot_forecast_vs_measured(
    forecasts_rain[209:377], measures_rain[209:377], 
    "Weekly Precipitation Forecast vs Measured", 
    "Hours since start of recording", "Precipitation (mm)", 
    "forecast_vs_real_rain_weekly.png"
)
print(linregress(forecasts_rain[209:377], measures_rain[209:377]))
```
